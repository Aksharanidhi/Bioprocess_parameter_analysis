```python
# Inspect the dataset
print (df.shape)
print(df.head())
df.info()
print(df.isna().sum())
print ("Duplicates:", df.duplicated().sum())
```

```python
# convert integer columns to float - uniform math operations.

int_cols = df.select_dtypes('int').columns
df[int_cols] = df[int_cols].astype(float)
```

```python
# Rename columns
df = df.rename(columns = {"Time (h)": "Time_h",
                         "Aeration rate(Fg:L/h)": "Aeration_L/h",
                         "Agitator RPM(RPM:RPM)": "Agitator_rpm",
                         "Sugar feed rate(Fs:L/h)": "Sugar_feed_L/h",
                          "Acid flow rate(Fa:L/h)": "Acid_flow_L/h",
                          "Base flow rate(Fb:L/h)": "Base_flow_L/h",
                          "Water for injection/dilution(Fw:L/h)": "WFI_dilution_L/h",
                          "Air head pressure(pressure:bar)": "Air_pressure_bar",
                          "Substrate concentration(S:g/L)" : "Substrate_conc_g/L",
                          "Dissolved oxygen concentration(DO2:mg/L)": "DO_concentration_mg/L",
                          "Penicillin concentration(P:g/L)": "Penicillin_conc_g/L",
                          "pH(pH:pH)": "pH",
                          "Temperature(T:K)": "Temperature_K",
                          "Oxygen Uptake Rate(OUR:(g min^{-1}))": "Oxygen_Uptake_Rate_g_min",
                          "Offline Penicillin concentration(P_offline:P(g L^{-1}))": "Offline_penicillin_conc_g/L",
                          "Offline Biomass concentratio(X_offline:X(g L^{-1}))" : "Offline_biomass_conc_g/L",
                          "Carbon evolution rate(CER:g/h)" : "Carbon_evol_rate_g/h",
                          "Viscosity(Viscosity_offline:centPoise)": "Offline_viscosity_cP"      
                         })
```

```python
# Since offline penicillin, biomass and viscosity has only 9 measurement - can't use them for model training.
# can't fill these missing values - since the data is recorded in offline. 
# Separate offline vs online columns - keep offline columns for later
```

```python
offline_cols = ["Offline_penicillin_conc_g/L", "Offline_biomass_conc_g/L", "Offline_viscosity_cP"]
online_df = df.drop(columns = offline_cols)   #use for EDA / model training
offline_df = df[["Time_h"] + offline_cols].dropna()
```

```python
print("online:", online_df.shape)
print("offline:", offline_df.shape)
```

```python
# check what correlates most with penicillin.

corr = df.select_dtypes(include=['number']).corr()
corr["Penicillin_conc_g/L"].sort_values(ascending=False)
```

```python
# Visualize correlation:

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure (figsize = (8,5))  # width = 8, height = 5
corr['Penicillin_conc_g/L'].sort_values(ascending = False).head(18).plot(kind = 'bar', color = "skyblue")
plt.title("Correlations with Penicillin Concentration")
plt.ylabel("Correlation Coefficient")
plt.show()
```

```python
# agitation does not correlate with any parameters since there is no variability.
# oxygen, DO, aeration, pressure - major predictors of penicillin production.
```

```python
# check correlation with online paramters only:
# Online parameters
corr_online = online_df.corr()
```

```python
penn_corr_online = corr_online['Penicillin_conc_g/L'] .sort_values(ascending = False)
print(penn_corr_online)
```

```python
plt.figure(figsize = (8,5))
sns.barplot(x=penn_corr_online[1:11].values, y=penn_corr_online[1:11].index, palette="coolwarm")
plt.title("Top Online Parameters Correlated with Penicillin Concentration")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Online Variable")
plt.show()
```

```python
# Scatterplot 1 : penicllin vs DO
plt.figure()
plt.scatter(online_df['DO_concentration_mg/L'], online_df['Penicillin_conc_g/L'])
plt.xlabel('DO concentration (mg/L)')
plt.ylabel('Penicillin concentration (g/L)')
plt.title('Penicillin vs DO')
plt.show()
```

```python
parameters = [
    'Time_h',
    'Carbon_evol_rate_g/h',
    'Air_pressure_bar',
    'Aeration_L/h',
    'Oxygen_Uptake_Rate_g_min',
    'DO_concentration_mg/L',
    'Sugar_feed_L/h',
    'WFI_dilution_L/h',
    'Base_flow_L/h',
    'pH',
    'Acid_flow_L/h',
    'Substrate_conc_g/L',
    'Temperature_K'
]

pen_conc = 'Penicillin_conc_g/L'

#create grid
rows = 5
cols = 3
plt.figure(figsize = (15,20))

for i, parameters in enumerate (parameters, 1):
    plt.subplot (rows, cols, i)
    plt.scatter(online_df[parameters], online_df[pen_conc])
    plt.xlabel(parameters, fontsize = 8)
    plt.ylabel ('penicillin concentration',fontsize = 8)
    plt.title (f'{parameters} vs Penicillin concentration', fontsize= 10)

plt.tight_layout()
plt.show()
```
